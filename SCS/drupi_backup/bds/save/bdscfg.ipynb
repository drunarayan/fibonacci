```python
#
#“source” 	(example: source = ‘cfls’)
#
#This string parameter sets the Base name of the spectral images obtained 
#and result files generated. These short single-word lower-case base names 
#will be appended with a timestamp as well as the extension denoting the 
#file types. These can then be downloaded to your laptop as needed using 
#the File menus from the Browser. The ‘source’ variable will need to be 
#changed when the source of the spectrum is switched. It is not necessary 
#to change it when doing multiple runs on the same source due to 
#appending the time stamp.

#
#“element” 	(example: element = ‘Mercury Terbium Argon’)
#
#Contains a single or multiple elements expected to be found in the source spectrum

#
#“desc” 	(example: desc = ‘Compact Fluorescent Lamp’)
#
#Contains the description of the source spectrum. Examples 
#are “Compact Fluorescent Lamp”, “Hydrogen Discharge Tube”, 
#“Copper Sulphate in Cuvette” etc.

#
#“shutter” 	(example: shutter = 100000)
#
#This integer parameter sets the value of the time in milliseconds that 
#the shutter will remain open. The range can vary from 10000 to 3000000. 
#The larger the number, the brighter the spectrum image.  It should be 
#kept to the least number possible where the spectral lines are barely 
#visible to the human eye. The camera can pick up spectral lines from 
#an almost “black” looking background. Over exposing the spectrum will 
#cause severe light leakage that will destroy the quality of the spectrum. 
#Very dim lines will require a bigger number for the shutter.

#
#wavelength_factor (example: wavelength_factor = 0.941)
#
#The wavelength_factor is the most important parameter for calibrating 
#a spectral line against a known standard. The small changes to wavelength 
#factor can dramatically change the spectral line distance from the slit 
#(measured wavelength). This number is calibrated to match the known wavelength 
#of Hydrogen or Mercury.  Once completed, the spectroscope usually does not 
#need to be calibrated for each run of a different element.
#This floating point integer value ranges between 0.70 and 0.99. It is usually 
#in this range to get a good calibration unless the diffraction grating is different 
#than the usual 1000 gratings per mm.
#wavelength_factor = 0.770 yielding green peak wavelength of Mercury around 
#470 nano-meters far away from the standard!
#wavelength_factor = 0.950 yielding green peak wavelength of Mercury around 
#570 nano-meters which is very close to the standard!

#
#“slit_topadj” 	(example: slit_topadj = 40)
#“slit_botadj” 	(example: slit_botadj = -40)
#
#The aperture is the rectangular boxed area drawn on the raw image of the spectrum. 
#This is the area that is used for pixel analysis. Spaces outside of this box 
#will be ignored.  This will help immensely when there is light leakage.
#This positive integer value is used to draw the top line of the rectangle. 
#The larger the positive value the lower the top line will be (more constrained) 
#This negative integer value for is used to draw the bottom line of the rectangle. 
#The larger the negative value the higher the bottom line will be (more constrained) 

#
#“spectrum_angle” 	(example: spectrum_angle = -0.01 or spectrum_angle = 0.1)
#
#This small floating point value adjusts the rotation of the rectangular aperture 
#drawn on the raw image. This needs to be horizontally oriented.
#Positive values rotate aperture anti-clockwise and Negative values clockwise

#
#samp_th = 0.4 wlen_th = 10 (skip peaks under 0.4)
#samp_th = 0.1 wlen_th = 10 (pick peaks under 0.4)
#samp_th = 0.1 wlen_th = 50 (do not pick more than one point within a 50-nano-meter width)
#
#There are 2 thresholds for peak wavelength determination Spectral Amplitude 
#Threshold and Wavelength Threshold.  The largger these numbers are the lesser 
#number of peaks that the gaussian calculation will find.


```
